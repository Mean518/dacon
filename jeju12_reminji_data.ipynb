```python
import pandas as pd
```

```python
min_data = pd.read_csv('min_data.csv',index_col=0)
min_data
```

```python
mapping = pd.read_csv('mapping.csv',index_col=0)
mapping
```

```python
mapping.columns=['STD_CLSS_NM','한국은행_분류']
```

```python
mapping
```

```python
min_data=pd.merge(min_data,mapping,how='left',on=['STD_CLSS_NM'])
```

```python
min_data
```

```python
서비스업_생산지수 = pd.read_csv('서비스업_생산지수.csv',index_col=0)
서비스업_생산지수
```

```python
a = min_data['REG_YYMM']
b = min_data['한국은행_분류']
```

```python
data = []
for i,j in zip(a,b) :
    ddd = 서비스업_생산지수.loc[i,j]
    data.append(ddd)
```

```python
min_data['서비스업_생산지수']=data
```

```python
min_data
```

```python
min_data.to_csv('min_data.csv')
```
