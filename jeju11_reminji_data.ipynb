```python
신용카드 = pd.read_csv('5.8.1 신용카드.csv', encoding='cp949', header=None).T
신용카드
```

```python
신용카드 = 신용카드.iloc[5:]
신용카드.columns = ['REG_YYMM','신용카드']
```

```python
신용카드['REG_YYMM']
```

```python
신용카드 = 신용카드.set_index
```

```python
신용카드
```

```python
data = []
for i in 신용카드['REG_YYMM'] :
    a = str(i[:4])+str(i[-2:])
    data.append(int(a))
```

```python
신용카드['REG_YYMM'] = data
```

```python
신용카드
```

```python
min_data
```

```python
min_data = pd.merge(min_data,신용카드,how='left',on=['REG_YYMM'])
```

```python
서비스업_생산지수 = pd.read_csv('16.5.1 산업별 서비스업생산지수.csv', encoding='cp949', header=0)
서비스업_생산지수
```

```python
서비스업_생산지수 = 서비스업_생산지수[서비스업_생산지수['항목명2']=='계절조정지수']
```

```python
del 서비스업_생산지수['통계표']
del 서비스업_생산지수['항목명2']
del 서비스업_생산지수['단위']
del 서비스업_생산지수['변환']
```

```python
서비스업_생산지수 = 서비스업_생산지수.T
```

```python
서비스업_생산지수.columns = 서비스업_생산지수.iloc[0]
```

```python
서비스업_생산지수 = 서비스업_생산지수.iloc[1:]
```

```python
서비스업_생산지수
```

```python
data = []
for i in 서비스업_생산지수.index :
    a = i[:4] + i[-2:]
    data.append(int(a))
```

```python
서비스업_생산지수.index = data
```

```python
서비스업_생산지수
```

```python
서비스업_생산지수.to_csv('서비스업_생산지수.csv')
```
